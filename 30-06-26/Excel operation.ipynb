### Change Data Types


```python
import pandas as pd

df = pd.read_excel("Student.xlsx")
print(df)
```

```python
df['Age'] = df["Age"].astype(str)
print(df.dtypes)
```

#### Date & Time

```python

df["Joining Date"] = pd.to_datetime(df["Joining Date"])
df["year"] = df["Joining Date"].dt.year
df["month"] = df["Joining Date"].dt.month_name()
df["day"] = df["Joining Date"].dt.day_name()

print(df[["Joining Date","year","month","day"]])
```

#### Fillter Data

```python
print("\n============ Marks > 80 ============")

result = df[df["Mark"] > 80]

print(result)
```

#### Sorting Data

```python
print("\n============================ Ascending ===============================")
result = df.sort_values("Mark")
print(result)

print("\n=========================== Descending ===============================")
result = df.sort_values("Mark", ascending=False)
print(result)
```

#### Add new Column

```python
df["Result"] = df["Mark"].apply(lambda x : "Pass" if x >= 33 else "Fail")
print(df)
```

#### Rename Column

```python
df.rename(
    columns={
        "Name": "Student_name"
    },
    inplace=True
)

print(df.columns)
print(df)

```

#### Delete Column

```python
df.drop("Age", axis=1, inplace=True)
print(df)
```

#### Save Updated Data into Excel

```python
df.to_excel("Student.xlsx", index= False)
print("\n Updated Data Saved in Excel File successfully")
```

#### Save Multiple Sheet

```python
with pd.ExcelWriter("output.xlsx") as writer:
    df.to_excel(writer, sheet_name="student_data", index=False)
```
