## Series in Python

#### A Series is a one-Dimansional labeled array capable of holding any Data type

```python
import pandas as pd
```

```python
data = pd.Series([])
print(data)
```

```python
data = pd.Series([1,2,3,4,5])
print(data)
```

### Series with custom index

```python
data = pd.Series(
    [90, 85, 80, 75, 70],
    index=["math", "science", "English", "special science", "Gujarati"]
)

print(data)
```

#### Access Series Data

```python
print(data["math"])
print(data["science"])
```

## DataFrame in python
#### A DataFrame is a two-Dimansional labeld tubular structure consider row and columns


```python
student = {
    "Name": ["Prarna", "Shrushti", "Puran", "Aditi", "Dixit"],
    "Age": [17, 16, 17, 17, 17],
    "City": ["Surat", "Ahemdabad", "Vadodra", "Jypur", "Surat"]  
    
}

print(student)
df = pd.DataFrame(student)

print(df)
```

### Creation of DataFrame

```python
data = [
    ["Rahul",20],
    ["Jay",19],
    ["Dixit",18]
]

df = pd.DataFrame(data)

print (df)
```

### Exploring DataFrame

```python
student = {
    "Name": ["Prarna", "Shrushti", "Puran", "Aditi", "Dixit"],
    "Age": [17, 16, 17, 17, 17],
    "City": ["Surat", "Ahemdabad", "Vadodra", "Jypur", "Surat"]  
    
}

df = pd.DataFrame(student)

print(df)
```

#### Display Last Row

```python
print(df.tail(1))

print(df.tail(-3))

print(df.dtypes)
```

### Shape of DataFrame

```python
print(df.shape)

print(df.columns)
```

### Statistical Information

```python
print(df.describe())

print(df.info())
```

## Operation on DaraFrame

#### Select single Column

```python
print(df["Name"])
```

#### Select Multiple column

```python
print(df[["Name","City"]])
```

Add new Column

```python
df ["Grade"]=['A', 'B', 'A', 'A', 'A']

print(df)
```

#### Update Column

```python
df["Age"]=df["Age"]+1
print(df)
```

#### Delete Column

```python
df.drop("Grade", axis=1, inplace=True)

print(df)
```

#### Filter Rows

```python
result = df[df["Age"]<20]
print(result)
```

### Sorting Data

1. Ascending
2. Decending

```python
print(df.sort_values("Age"))

print(df.sort_values("Age", ascending=False))

print(df["Age"].max())

print(df["Age"].min())

print(df["Age"].mean())

print(df["Age"].sum())
```

#### Pandas Series All types with NaN value

### What is NaN ?
#### NaN [Not A Number] represent missing or undefined values in pandas 


```python
import numpy as np

s = pd.Series([10,20,np.nan,40,50])
print(s)    
```

```python
s = pd.Series(["Amit", "Dixit", "Maru", "Python", "Pandas"])
print(s)

s = pd.Series([True, False, True, False, True])
print(s)    

s = pd.Series([1,2,3,4,5], index=["a", "b", "c", "d", "e"])
print(s)            


```

### Series With None


```python
s = pd.Series([10,20,np.nan,40,5])
print(s)
print(s.isnull())
print(s.notnull())
```

### Fill NaN values
#### Using fillna()

```python
s = pd.Series([10, np.nan, 30, 40])

s = s.fillna(0)

print(s)
```

Using Dropna()

```python
s = pd.Series([10 ,np.nan ,30 ,40 ])

print(s.dropna())
```
