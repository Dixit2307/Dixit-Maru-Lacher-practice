## Create,Read and Write CSV,Excel & JSON

```python
import pandas as pd
```

### Create Dictionary

```python
student_data = {
    "Name":["Vivak","Rahul","Naha","Priya","Zeel"],
    "Age":[28,25,21,23,50],
    "Mark":[89,78,91,None,89],
    "City":["Surat","Ahemdabad","Vadodra",None,"Rajkot"],
    "Joining Date":["15-06-2026","20-05-2025","20-06-2024","01-02-2023","01-05-2026"]
}

df=pd.DataFrame(student_data)
print(df)
```

### to_csv (For creating CSV File)

```python
df.to_csv("Student.csv", index=False)
print("\nCSV File Created successfully")
```

### to_excel (For Creating Excel File)

```python

df.to_excel("Student.xlsx", index=False)
print("Excel File Created Successfully")
```

### to_JSON (For Create json File)

```python
df.to_json("Student_data.json",orient="records",index=4)
print("File created successfully")
```

### Read csv File

```python
csv_data = pd.read_csv("student.csv")
print(csv_data)

print("\n File Read Successfully")
```
