## Pandas Resd and Manipulate one or multiple excle sheets

```python
import pandas as pd
```

```python
print("========Read First Sheet========")

df = pd.read_excel("Student.xlsx")
print(df)
```

#### Read Spacific sheet

```python
sheet2 = pd.read_excel("Student.xlsx", sheet_name="Sheet2")
print(sheet2)
```

#### Read Sheet by index

```python
sheet2 = pd.read_excel("Student.xlsx", sheet_name=1)
print(sheet2)
```

#### Read Multiple sheet

```python
print("======== Read Multiple Sheet ========")

multiple_sheet = pd.read_excel("Student.xlsx",sheet_name=[0,1])
print(multiple_sheet)
```

#### Read All Sheets 

```python
print("========Read all sheet ========")

all_sheet = pd.read_excel("Student.xlsx", sheet_name = None)
for sheet_name , data in all_sheet.items():
    print("Sheet Name", sheet_name)
    print("====================================================")
    print(data)
```

#### Select Data

```python
print("\n ===== Single Column =====")
print(df["Name"])
```

```python
print("\n===== Multiple Column =====")
print(df[["Name","Mark"]])
```

### Missing Values

```python
print("\n ====== Missing Values ======")
print(df.isnull().sum())
df["Mark"]=df["Mark"].fillna(df["Mark"].mean())
df["City"]=df["City"].fillna("Unknown")
print(df)
```

### Remove Duplicate recores

```python
print("\n ======= Duplicate Rows ======")
df.duplicated().sum()
df.drop_duplicates()
print("\n====== after Duplicate ======")
print(df)
```
